```python
import pandas
import pyLDAvis
import pyLDAvis.lda_model
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation         # 분장에서 토픽을 찾을수 있다.
from sklearn.datasets import fetch_20newsgroups
# 데이터 로드
data = fetch_20newsgroups(subset='train', remove=('headers', 'footers', 'quotes'))
documents = data.data
```

```python
documents[:1]
```

### TfidfVectorizer
- max_df= : 단어가 전체 문서의 95% 이상에서 나타나면 해당 단어는 무시해라.(너무 많이 반복되는 단어(너무 too much하게 반복이 많으면 의미가 없다고 봄.))
- min_df= 단어가 최소한 2개의 문서에서 나타나야만 해당 단어가 유지해라.(너무 극소수로 나타나는 단어(너무 적은 수의 단어가 의미가 특별히 없다고 봄.))

```python
tfidfVectorizer = TfidfVectorizer(stop_words='english', max_df=0.95, min_df=2)
```

```python
tfidf = tfidfVectorizer.fit_transform(documents)
tfidf.toarray()[:1]
```

```python
import numpy as np
np.where(tfidf.toarray()[:1] > 0)
```

## 5개로 범주화

```python
latentDirichletAllocation = LatentDirichletAllocation(n_components=5, n_jobs=-1)     # n_components=5 : 5개로 범주화하라는 코드
```

```python
latentDirichletAllocation.fit(tfidf)
```

```python
vis = pyLDAvis.lda_model.prepare(latentDirichletAllocation, tfidf, tfidfVectorizer)
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.display(vis)
```
