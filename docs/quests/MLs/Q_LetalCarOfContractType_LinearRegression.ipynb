## 데이터

```python
import pandas as pd
df_RCOCT = pd.read_csv('/content/RentalCarOfContractType.csv')
df_RCOCT[:2]
```

```python
# 컬럼에서 결측치가 유무 확인
df_RCOCT.isnull().sum()
```

## feature, target 선택
- target : age
- featuru : Term, amount

```python
# 컬럼 추출
# df_RCOCT.loc[:,['age', 'Term', 'amount']][:2]
df_RCOCT_extract = df_RCOCT.loc[:,['Term', 'amount']]
df_RCOCT_extract[:2]
```

```python
# 결측치와 이상치 확인 -> 공백 여부 확인 필요
df_RCOCT_extract.isnull().sum()
```

## 데이터 분리
- target과 labels(festures) 분리
- train과 test set 분리

```python
target = df_RCOCT_extract['age']
features = df_RCOCT_extract[['Term', 'amount']]
target.shape, features.shape
```

```python
from sklearn.model_selection import train_test_split
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.3, random_state=2)
features_train.shape, features_test.shape, target_train.shape, target_test.shape
```

## 모델

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# 학습 진행
model.fit(features_train, target_train)
```
